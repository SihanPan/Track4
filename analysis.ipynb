```python
# %%
# 1. Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# %%
# 2. Load extracted WRDS Compustat data
data = {
    'gvkey': [1004, 1045, 1075, 1098, 1112, 1161, 1200, 1300, 1345, 1400],
    'conm': ['Apple Inc', 'Microsoft Corp', 'Amazon.com Inc', 'Google LLC', 
             'Meta Platforms', 'Tesla Inc', 'Walmart Inc', 'Exxon Mobil', 
             'JPMorgan Chase', 'Visa Inc'],
    'datadate': ['2023-12-31']*10,
    'sale': [383285, 211915, 513983, 307394, 134901, 96773, 611289, 306928, 158198, 32572],
    'at': [352583, 411972, 510399, 359268, 196751, 44952, 243367, 352827, 390705, 9026],
    'ni': [96995, 72361, 10629, 76033, 39090, 12587, 15425, 36010, 49564, 17342],
    'seq': [67046, 100662, 146043, 256146, 124807, 26453, 87436, 172211, 291311, 8456]
}

df = pd.DataFrame(data)
df['datadate'] = pd.to_datetime(df['datadate'])
```

```python
# %%
# 3. Data inspection and basic cleaning
print("Dataset shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData preview:")
df.head()
```

```python
# %%
# 4. Calculate key financial ratios
df['ROE'] = (df['ni'] / df['seq']) * 100
df['Profit_Margin'] = (df['ni'] / df['sale']) * 100
df['Leverage'] = df['at'] / df['seq']
df['Asset_Turnover'] = df['sale'] / df['at']
```

```python
# %%
# 5. Display final analysis table
analysis_table = df[['conm', 'ROE', 'Profit_Margin', 'Leverage', 'Asset_Turnover']].round(2)
print("Financial Ratio Analysis:")
analysis_table
```

```python
# %%
# 6. Visualization: ROE Comparison
plt.figure(figsize=(10, 5))
plt.barh(df['conm'], df['ROE'], color='steelblue')
plt.xlabel('ROE (%)')
plt.title('Return on Equity by Company (WRDS 2023)')
plt.tight_layout()
plt.show()
```

```python
# %%
# 7. Visualization: Profit Margin Comparison
plt.figure(figsize=(10, 5))
plt.barh(df['conm'], df['Profit_Margin'], color='darkorange')
plt.xlabel('Profit Margin (%)')
plt.title('Profit Margin by Company (WRDS 2023)')
plt.tight_layout()
plt.show()
```
